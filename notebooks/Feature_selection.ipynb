```python
#importing libraries
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import RidgeCV, LassoCV, Ridge, Lasso
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('TrainingWiDS2021.csv')
dum_df = pd.get_dummies(df[["ethnicity",'gender','hospital_admit_source','icu_admit_source','icu_stay_type','icu_type']], columns=["ethnicity",'gender','hospital_admit_source','icu_admit_source','icu_stay_type','icu_type'] )
bridge_df = df.join(dum_df)
bridge_df = bridge_df.drop(["ethnicity",'gender','hospital_admit_source','icu_admit_source','icu_stay_type','icu_type'],1)
bridge_df = bridge_df.fillna(bridge_df.median())
X = bridge_df.drop("diabetes_mellitus",1)   #Feature Matrix
y = bridge_df["diabetes_mellitus"]  
```

```python
#Using Pearson Correlation
plt.figure(figsize=(160,160))
cor = df.corr()
sns.heatmap(cor, cmap=plt.cm.Reds,)
plt.show()
```

```python
#Correlation with output variable
cor_target = abs(cor["diabetes_mellitus"])
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.5]
relevant_features
```

```python
X_1 = sm.add_constant(X)
#Fitting sm.OLS model
model = sm.OLS(y,X_1).fit()
model.pvalues
```

```python
#Backward Elimination
cols = list(X.columns)
pmax = 1
while (len(cols)>0):
    p= []
    X_1 = X[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(y,X_1).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```
